```python
import sys
sys.path.append('../modules')
from wrangling import wrangling_indices as wr
from grafo_estacionalidad import Grafo_Estacionalidad as graf
import pandas as pd
```

```python
fechaCorrida = 'Marzo 2022'

expo = graf(wr(path='../data/marzo2022/serie_mensual_indices_comex.xls'),
            'iq_x',
            'ip_x')
impo = graf(wr(path='../data/marzo2022/serie_mensual_indices_comex.xls'),
            'iq_m',
            'ip_m')


expo.set_title('ndices de exportacin - Enero 2011 a Marzo 2022\n')
expo.set_table_title(['Cantidades', 'Precios'])
expo.set_size(20,15)
expo.hacer_grafo()
expo.set_legends('Cantidades', 'Precios')
expo.set_ylabel('ndice base = 2004')

impo.set_title('ndices de importacin - Enero 2011 a Marzo 2022\n')
impo.set_table_title(['Cantidades', 'Precios'])
impo.set_size(20,15)
impo.set_colors('darkred', 'darkblue', 'darkblue', 'darkred')
impo.hacer_grafo()
impo.set_legends('Cantidades', 'Precios')
impo.set_ylabel('ndice base = 2004')
```

```python
expo.print(f'../output/indices-PrecioCantidad-Expo({fechaCorrida}).pdf')
impo.print(f'../output/indices-PrecioCantidad-Impo({fechaCorrida}).pdf')
```

```python
impo.df_tabla().to_excel(f'../output/tablas/indices-PrecioCantidad-Impo({fechaCorrida}).xlsx')
impo.df_tabla().to_excel(f'../output/tablas/indices-PrecioCantidad-Expo({fechaCorrida}).xlsx')
```
