```python
import sys
sys.path.append('../modules')
from wrangling import wrangling_indices as wr
from grafo_estacionalidad import Grafo_Estacionalidad as graf
import pandas as pd
```

```python
wr()
```

```python

fechaCorrida = 'Marzo 2022'

expo = graf(wr(path='../data/marzo2022/serie_mensual_indices_comex.xls'),
            'iq_x',
            'ip_x')
impo = graf(wr(path='../data/marzo2022/serie_mensual_indices_comex.xls'),
            'iq_m',
            'ip_m')

iti =  graf(wr(path='../data/marzo2022/serie_mensual_indices_comex.xls'),
            'ITI')

iti.set_title('Índice de términos de intercambio. Enero 2011 - Marzo 2022')
iti.set_size(20,10)
iti.set_colors('#4f4561', '#cbade0')
iti.set_lineWidth(4)

iti.hacer_grafo()
iti.set_ylabel('Índice base = 2004')

expo.set_title('Índices de exportación - Enero 2011 a Marzo 2022\n')
expo.set_table_title(['Cantidades', 'Precios'])
expo.set_size(20,10)
expo.set_colors('#edc24c','darkred', '#ffe294', 'darkred')
expo.set_lineWidth(4)
expo.set_ticksize(20)
expo.hacer_grafo()
expo.set_legends('Cantidades', 'Precios', size = 25)
expo.set_ylabel('Índice base = 2004')

impo.set_title('Índices de importación - Enero 2011 a Marzo 2022\n')
impo.set_table_title(['Cantidades', 'Precios'])
impo.set_size(20,10)
impo.set_colors('#9885a5', 'darkred', '#cbade0', 'darkred')
impo.set_lineWidth(4)
expo.set_ticksize(20)
impo.hacer_grafo()
impo.set_legends('Cantidades', 'Precios', location=2, size= 25)
impo.set_ylabel('Índice base = 2004')

```

```python
expo.print(f'../output/indices-PrecioCantidad-Expo({fechaCorrida}).pdf')
impo.print(f'../output/indices-PrecioCantidad-Impo({fechaCorrida}).pdf')
```
