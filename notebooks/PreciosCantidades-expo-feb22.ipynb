```python
import sys
sys.path.append('../modules')
from wrangling import wrangling_indices as wr
from grafo_estacionalidad import Grafo_Estacionalidad as graf
```

```python
expo = graf(wr(),
            'iq_x',
            'ip_x')
impo = graf(wr(),
            'iq_m',
            'ip_m')


expo.set_title('Índices de exportación - Enero 2011 a Febrero 2022\n')
expo.set_table_title(['Cantidades', 'Precios'])
expo.set_size(20,20)
expo.hacer_grafo()
expo.set_legends('Cantidades', 'Precios')
expo.set_ylabel('Índice base = 2004')

impo.set_title('Índices de importación - Enero 2011 a Febrero 2022\n')
impo.set_table_title(['Cantidades', 'Precios'])
impo.set_size(20,20)
impo.set_colors('darkred', 'darkblue', 'darkblue', 'darkred')
impo.hacer_grafo()
impo.set_legends('Cantidades', 'Precios')
impo.set_ylabel('Índice base = 2004')
```

```python
expo.print('../output/indices-PrecioCantidad-Expo.pdf')
impo.print('../output/indices-PrecioCantidad-Impo.pdf')
```
